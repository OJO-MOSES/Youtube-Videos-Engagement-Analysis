```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\OJO MOSES\Desktop\COPY\CAvideos.csv")
df.head()
```

```python
#Checking the data size using the .shape function
df.shape
```

```python
#Checking the descriptive analysis of the dataset
df.describe()
```

```python
#Checking general information in the dataset using the .info() functio
df.info()
```

```python
# Convert publish_time to datetime
df['publish_time'] = pd.to_datetime(df['publish_time'])

# Convert trending_date to formated datetime to elimate erros in time zone
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
```

```python
df.info()
```

```python
# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# If there are duplicate rows, you can drop them using the following code:
df.drop_duplicates(inplace=True)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Fill missing values in the description column with an empty string
df['description'].fillna('', inplace=True)
```

```python
#Rechecking our non values
df.isnull().sum()
```

```python
new_df = df.copy()
new_df.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))  # Adjust the width (12) and height (6) as needed
top_views = new_df.nlargest(10, 'views')
plt.bar(top_views['title'], top_views['views'])
plt.xlabel('Video Title')
plt.ylabel('Views')
plt.title('Top 10 Most Viewed Videos')
plt.xticks(rotation=90, ha='right', fontsize=8, wrap=True)
# plt.gca()xaxis.set
plt.tight_layout()
plt.show()
```

```python
# we wiill now derive a metric to quantify the engagement rate for each video,
# which will enable us to better understand how users are interacting with our content."
new_df['engagement_rate'] = (new_df['likes'] + new_df['dislikes']) / new_df['views']
new_df['engagement_rate'].describe()
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram for views
axes[0].hist(new_df['views'], bins=20)
axes[0].set_xlabel('Views')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of Views')

# Histogram for engagement rate
axes[1].hist(new_df['engagement_rate'], bins=20)
axes[1].set_xlabel('Engagement Rate')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Distribution of Engagement Rate')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# The distribution of videos across different categories to identify which ones have the highest volume of content."
category_counts = new_df['category_id'].value_counts()
category_counts
```

```python
# Conducting a comparative analysis of YouTube channel video repositories.
# By extracting and quantifying the video datasets to determine which channel possesses the largest collection
channel_counts = new_df['channel_title'].value_counts()
channel_counts
```

```python
top_channels = channel_counts.nlargest(10)
plt.bar(top_channels.index, top_channels.values)
plt.xlabel('Channel Title')
plt.ylabel('Number of Videos')
plt.title('Top 10 Channels with Most Videos')
# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=90, ha='right', fontsize=10, wrap=True)
plt.tight_layout()
plt.show()
```

```python
# Checking for the time videos are most commonly published:
publish_time_counts = new_df['publish_time'].dt.hour.value_counts()
publish_time_counts
print("We can therefore see that most videos are posted around 16:00 hours of the day.")
```

```python
plt.figure(figsize=(12, 6))  # increase figure size
plt.bar(publish_time_counts.index, publish_time_counts.values)
plt.xlabel('Hour of Day')
plt.ylabel('Number of Videos Published')
plt.title('Videos Published by Hour of Day')

# Show all 24 hours on the x-axis with improved spacing
plt.xticks(range(24), [f'{h:02d}:00' for h in range(24)], rotation=90, ha='right', fontsize=10)

colors = plt.cm.tab20(range(20))  # use 20 distinct colors from the tab20 colormap

plt.bar(publish_time_counts.index, publish_time_counts.values, color=[colors[i % len(colors)] for i in range(len(publish_time_counts))])

plt.tight_layout()  # ensure labels fit within the figure area
plt.show()
```

```python
#Rate of video trending by day
trending_date_counts = new_df['trending_date'].dt.dayofweek.value_counts()
trending_date_counts
```

```python
# Graph showing the rate of video trending by day
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.bar(days_of_week, trending_date_counts.values)
plt.xlabel('Day of Week')
plt.ylabel('Number of Videos Trending')
plt.title('Videos Trending by Day of Week')
plt.xticks(rotation=45, fontsize=12)
plt.show()
```

```python
#Correlation between views, likes, dislike.
corr_matrix = new_df[['views', 'likes', 'dislikes']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```
